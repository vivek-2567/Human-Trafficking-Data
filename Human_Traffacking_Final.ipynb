```python
def make_plots(df, number,dimension = (20,20)):
    
    position = []
    counter = 0

    if number%2==0:

        iterator = int(number/2)
        for i in range(iterator):
            for j in range(2):
                position.append([i,j])

    else:
        iterator = int((number+1)/2)
        for i in range(iterator):
            for j in range(2):
                position.append([i,j])

    plt.rcParams["figure.figsize"] = dimension

    df_columns = df.columns

    figure, axis = plt.subplots(iterator, 2)

    for i in range(len(df_columns)):

        try:
            x = df.nlargest(5,df_columns[i])
            axis[position[counter][0],position[counter][1]].plot(x[df_columns[0]],x[df_columns[i]])
            titles = str(df_columns[i]).capitalize()
            axis[position[counter][0],position[counter][1]].set_title(titles)
            counter += 1

        except:
            continue
        
    plt.plot()
```

```python
def make_plots_yrs(df, df1, df2, number,dimension = (20,20)):
    
    counter = 0

    plt.rcParams["figure.figsize"] = dimension

    df_columns = df.columns
    df1_columns = df1.columns
    df2_columns = df2.columns

    figure, axis = plt.subplots(number, 3)

    for i in range(len(df_columns)):

        try:
            x = df.nlargest(5,df_columns[i])
            axis[counter,0].plot(x[df_columns[0]],x[df_columns[i]])
            titles = str(df_columns[i]).capitalize() + " in 2018"
            axis[counter,0].set_title(titles)

            x1 = df1.nlargest(5,df1_columns[i])
            axis[counter,1].plot(x1[df1_columns[0]],x1[df1_columns[i]])
            titles = str(df1_columns[i]).capitalize()  + " in 2019"
            axis[counter,1].set_title(titles)

            x2 = df2.nlargest(5,df2_columns[i])
            axis[counter,2].plot(x2[df2_columns[0]],x2[df2_columns[i]])
            titles = str(df2_columns[i]).capitalize() + " in 2020"
            axis[counter,2].set_title(titles)

            counter += 1

        except:
            continue
        
    plt.plot()
```

```python
make_plots_yrs(culprits_18,culprits_19,culprits_20,4)
```

4. Person acquitted/discharged by the court

    ii. **Telengana and Andra Pradesh** holds 2nd and 3rd postion, sometimes interchanging the position between themselves.

3. Domestic survitude

    i. The number of people employed in this is increasing every year.

    iii. **Jharkhand and Andra Pradesh** remains in top 5 for all 3 concucative years.

4. Forced Marriage

    i. **Assam, Kerala, Madhya Pradesh, Uttar Pradesh and West Bengal** reamins in the top 5 for this cause and just interchange the position in the 3 consucative years.

    ii. Number of people involved in this decreased in 2020 (covid year) as compared to 2018 and 2019.

5. Petty Crimes

    i. Not a lot of people are involves in this.

    ii. Generally less than 10 people are trafficed for this cause.

6. Child Pornography

    i. Accoring to the data, only **Karnataka** state has this problem.
    ii. In 2019, 0 cases found where people are kidnapped for this cause.

7. Begging

    i. **Kerela** is the the state seen in the top 5 list for this activity for 3 consecutive years.

    ii. In 2019, the country witnessed the most number of people trafficed for this reason from a single state.

8.  Removal of organs

    i. Not a lot of people are trafficked for this activity.

    ii. The maximum number of people trafficked are generally less than 10.

    iii. In 2020, We saw no cases related to this activity in the whole country.

9. Other Purposes

    i. **Odisha** remains in the top 5 list for 3 consucative years.

    ii. People trafficked for other purposes peaked in 2019 as compared to 2018 and 2020.
